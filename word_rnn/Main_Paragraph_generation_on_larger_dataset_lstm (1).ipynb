```python
# c is just and intermediatory step in formatting data. for loop will remove unnecessary end spaces in each paragraph.
c = []
for i in range(len(b)):
    b[i] = b[i].strip()
```

Following block will replace unnecessary \n characters present in the dataset and convert all data into lower case. 

Since many paragraphs were useless  like "PAGE {} | BOOk1...." those need to be removed.

```python
flag=0
for i in range(1,len(b)-1):
    if b[i].isupper()==0 and flag<=0:             ### it is observed at some places, headings of chapter were UPPER case sentences. so ignore those. 
        if b[i].find('Page | ')==0:
            if b[i-1] != '' and b[i-1][-1] != '.':           ###  Join previous and next paragraph if current paragrgaph represents end of page 
                temp = " ".join([b[i-1],b[i+1]])
                c.append(temp.replace('\n','').lower())
                flag = 1
        elif flag<=0:
            c.append(b[i].replace('\n','').lower())
    flag-=1
```

```python
### Even after removing page end sentences in above block, some were left. Those are removed here.
for i in range(len(c)):
    ind = c[i].find('page | ')
    if ind!=-1:
        c[i] = c[i][:ind]
```

```python
### d will be final formatted list that contains all paragraphs. In paragraphs in list c, some inecessary spaces were also added. those will be removed now.
d = []
for i in c:
    if i!='':
        d.append(i)
```

```python
data2 = []   ### this will be our train iterator.

for i in d:
    a = []
    temp = i.strip()
    word = ''
    j = 0
    while j<len(temp):      ### iterate over every character in paragraph and form seperate words and punctuation marks and remove unnecesary spaces.
        if temp[j]==' ':      
            a.append(word.lower())    ## to make whole data in lower case alphabets
            word = ''
        elif temp[j] in punctuation:
            if word != '':
                a.append(word.lower())
            a.append(temp[j])
            word = ''
            
        else:
            word += temp[j]
        j +=1 
    if len(word) != 0:
        a.append(word.lower())
    if a[-1] != '\n':
        a.append('\n')            ## \n will be our end token. it will represent end of paragraph. Hence add it if not present.
    data2.append(a)
```

### Helper function

Following encodings2 function tokenizes paragraphs based on word_dict2

```python
def encodings2(sentence,word_dict2):
    vocab_size = len(word_dict2)
    input_size = len(sentence)
    one_hot_sentence = torch.zeros([input_size-1],dtype=torch.long)
    target = torch.zeros([input_size-1],dtype=torch.long)

    for i in range(len(sentence)-1):
        word = sentence[i]
        next_word = sentence[i+1]
        one_hot_sentence[i] = word_dict2[word]
        target[i] = word_dict2[next_word]
    #print(one_hot_sentence.shape,target.shape)
    return(one_hot_sentence,target)
```

In RNN class you can change LSTM layer with RNN or GRU layers

## **Creating Model**

Stanford CS224N_ NLP with Deep Learning _ Winter 2019 _ Lecture 6 – Language Models and RNNs - YouTube - Google Chrome 04_05_2021 2_18_30 PM.png

```python
## This followes simple language modeling architecture mentioned in the cs224n lectures.
## Here, code is written for RNN constructed with LSTM units.
## If you are using GRU or RNN, then do not initialize and pass cell state in nn.GRU/nn.RNN. 
## Change code accordingly.
class RNN(nn.Module):
    def __init__(self,hidden_size,output_size,vocab_size,embedding,num_layers):
        super(RNN,self).__init__()
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.num_layers = num_layers
        self.emb = embedding
        self.embedding = nn.Embedding(vocab_size,embedding)            ## Embedding layer 1st layer.
        self.lstm = nn.LSTM(embedding,hidden_size,batch_first=True,num_layers=num_layers)    ## RRN layer 2nd layer
        self.dropout = nn.Dropout(0.2)        ## 0.2 dropout was pretty much usefull to avoid overfitting.
        self.fc = nn.Linear(hidden_size,output_size)            ## output 3rd layer
    def forward(self,x,hidden,cell):
        x = self.embedding(x)
        x = x.reshape(1,x.shape[0],x.shape[1])
        x,(hidden,cell) = self.lstm(x,(hidden,cell))
        x = self.dropout(x)
        x = self.fc(x)
        return(x,(hidden,cell))
    def initHidden(self):           ## creates and returns hidden layer. Zero hidden layer works better that random initialization.
        return(torch.zeros([self.num_layers,1,self.hidden_size]))
```

*Here we havent used Softmax layer because further we will be using crossentropyloss which will be haing that function*

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
embedding = 128
output_size = vocab_size = len(word_dict2)
hidden_size = 256
num_layers = 4
learning_rate = 0.0006
epochs = 25
```

```python
## Initialize model, optimizer, criterion.
## We are using here Adam's optimizer since it gives optimum performannce.
model2 = RNN(hidden_size,output_size,vocab_size,embedding,num_layers).to(device)
optimizer = optim.Adam(model2.parameters(),lr=learning_rate)
criterion = nn.CrossEntropyLoss()
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f'The model has {count_parameters(model2):,} trainable parameters')
```

## **Training**


---
 Training takes approx 25 sec per epochs and run for 25 epochs to get best 
results.

```python
loss_lst = []
for epoch in range(epochs):  # epochs
    epoch_loss = []
    t1 = time.time()
    for sentence in data2:
        if(len(sentence)==1):
            continue
        sentence_encoding,target = encodings2(sentence,word_dict2)
        sentence_encoding = sentence_encoding.to(device)
        target = target.to(device)
        #print(sentence_encoding)
        #print(target.shape)
        hidden = model2.initHidden().to(device)
        cell = model2.initHidden().to(device)
        #print(cell.shape)
        model2.zero_grad()
        optimizer.zero_grad()

        out,(hidden,cell) = model2(sentence_encoding,hidden,cell)
        #print(out.shape)
        out = out.reshape(-1,output_size)
        loss = criterion(out,target)
        epoch_loss.append(loss.item())
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model2.parameters(), 1)     # gradient clipping prevents model from exploading gradients.
        optimizer.step()
    loss_lst += epoch_loss
    t2 = time.time()
    print("mean epoch loss : {}  ||  mean loss till now : {}  ||  time : {}".format(np.mean(epoch_loss),np.mean(loss_lst),t2-t1))
```

```python
## Plotting loss function. Here we have sampled loss after pass of every paragraphs. Thats why you may see such king of curve with highs and lows.
## You can apply any gaussian filter to get a smooth curve.
plt.plot(loss_lst)
```
